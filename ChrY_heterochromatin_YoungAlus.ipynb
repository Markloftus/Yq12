## Find the Young Alus

```python
def pullYoungAlus(df):
    tempList={}
    for row in df.index:
        for unit in ast.literal_eval(df.at[row,'Group']):
            if 'SINE/Alu' in unit:
                if float(unit.split(" ")[2]) < 10.0:
                    tempList[row]=unit
                else:
                    continue
    return(tempList)
```

```python
def pullYoungAlusT2T(df):
    tempList={}
    for row in df.index:
        for unit in ast.literal_eval(df.at[row,'Group']):
            if 'SINE/Alu' in unit:
                if float(unit.split(" ")[1]) < 10.0:
                    tempList[row]=unit
                else:
                    continue
    return(tempList)
```

```python
def buildDrawBlock(df, dictt, graveStart, graveEnd):
    dictt2 = dictt.copy()    
    HetLength = graveEnd-int(df.iloc[0]['Start'])

    finalDict = {x:[] for x in df.index}
    for alu in dictt2.keys():
        
        for row in df.index:
            if int(alu.split("_")[2].split("-")[0])>= int(df.at[row,'Start']) and int(alu.split("_")[2].split("-")[1])<= int(df.at[row,'End']):
                finalDict[row].append(alu)
            else:
                continue
                
                
    tempList = []
    StartPoint=0.0
    for row in df.index:
        if len(finalDict[row]) > 0:
            
            startPosition = int(df.at[row,'Start'])
            
            flag=0
            for alu in finalDict[row]:
                                
                aluStart = int(alu.split("_")[2].split("-")[0])
                aluEnd = int(alu.split("_")[2].split("-")[1])
                
                
                if flag == 0:
                    if df.at[row,'Block_Type'] == 'Repeat Block':
                        tempList.append([df.at[row,'Block_Number'], 'Repeat Block w/Alu', startPosition, aluStart-1, df.at[row,'SampleName'], '#D81B60'])
                        flag+=1
                    else:
                        tempList.append([df.at[row,'Block_Number'], 'Spacer Block w/Alu', startPosition, aluStart-1, df.at[row,'SampleName'], '#FFC107'])
                        flag+=1
                else:
                    if df.at[row,'Block_Type'] == 'Repeat Block':
                        tempList.append([df.at[row,'Block_Number'], 'Repeat Block w/Alu', startPosition+1, aluStart-1, df.at[row,'SampleName'], '#D81B60'])
                        flag+=1
                    else:
                        tempList.append([df.at[row,'Block_Number'], 'Spacer Block w/Alu', startPosition+1, aluStart-1, df.at[row,'SampleName'], '#FFC107'])
                        flag+=1

                if df.at[row,'Block_Type'] == 'Repeat Block':
                    tempList.append([df.at[row,'Block_Number'], 'Repeat Block w/Alu', aluStart, aluEnd, df.at[row,'SampleName'], '#D81B60'])
                else:
                    tempList.append([df.at[row,'Block_Number'], 'Spacer Block w/Alu', aluStart, aluEnd, df.at[row,'SampleName'], '#FFC107'])
                startPosition=aluEnd
                
                
            if int(df.at[row,'End'])-startPosition ==0:
                continue
            else:
                newStart = startPosition+1
                
                if df.at[row,'Block_Type'] == 'Repeat Block':
                    tempList.append([df.at[row,'Block_Number'], 'Repeat Block w/Alu', newStart,int(df.at[row,'End']), df.at[row,'SampleName'], '#D81B60'])

                else:
                    tempList.append([df.at[row,'Block_Number'], 'Spacer Block w/Alu', newStart,int(df.at[row,'End']), df.at[row,'SampleName'], '#FFC107'])

            
        else:
            if df.at[row,'Block_Type'] == 'Repeat Block':
                tempList.append([df.at[row,'Block_Number'], df.at[row,'Block_Type'], int(df.at[row,'Start']),int(df.at[row,'End']), df.at[row,'SampleName'], '#004D40'])

            else:
                tempList.append([df.at[row,'Block_Number'], df.at[row,'Block_Type'], int(df.at[row,'Start']), int(df.at[row,'End']), df.at[row,'SampleName'], '#D8D8D8'])
    
    
    tempList.append(['NA', 'Non-Analyzed', graveStart, graveEnd, df.at[row,'SampleName'], '#100F0F'])

    
    tempDF = pd.DataFrame(data=tempList, columns=['Block_Number','Label','start','end','SampleName', 'color_hex'])
    
    #new_row = pd.DataFrame({'Block_Number':'na', 'Label':'Non-Heterochromatin', 'start':0,
                        #'end':int(tempDF.iloc[0]['start'])-1, 'SampleName':tempDF.iloc[0]['SampleName'], 'color_hex':'#ADD8E6'},index =[0])
    
    #tempDF = pd.concat([new_row, tempDF]).reset_index(drop = True)
    print(finalDict)
    return(tempDF)
    
```
